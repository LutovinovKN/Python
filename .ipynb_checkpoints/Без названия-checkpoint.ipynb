```python
from sympy import *
```

```python
x = Symbol('x', real = True)
```

```python
y = ((-12*x)**4)*(sin(cos(x)))-((18*x)**3)+((5*x)**2)+(10*x)-30
```

```python
plot(y)
```

```python
solve(y, x)
```
